# Teacher Compensation Survey (pandas practice)
This notebook is meant as practice in cleaning data and tables in Python. The data here is from the Teacher Compensation Survey obtained from https://nces.ed.gov/ccd/tcssurv.asp.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from urllib.request import urlretrieve
from shutil import unpack_archive
```

```python
#Downloading zip files with data and then extracting them.
url1 = 'https://nces.ed.gov/ccd/data/zip/Tcs_sch_2006071a_txt.zip'
url2 = 'https://nces.ed.gov/ccd/data/zip/tcs_sch_2005061a_txt.zip'

urlretrieve(url1, 'Tcs_sch_2006071a_txt.zip')
urlretrieve(url2, 'tcs_sch_2005061a_txt.zip')

unpack_archive('Tcs_sch_2006071a_txt.zip')
unpack_archive('tcs_sch_2005061a_txt.zip')
```

## School Year 2005-06

```python
#School-Level Teacher Compensation Data 2005-06: Importing data from tab-separated flat file
tcs_sch_200506 = pd.read_csv('tcs_sch_2005061a.txt', sep='\t')

#Checking a sample of the table
tcs_sch_200506.sample(5)
```

Notice that some of the columns are float values and others are integers, but all columns except for the first one are supposed to be floats.

```python
#Reviewing the types of each column
tcs_sch_200506.dtypes
```

Per the documentation, the NCES School ID (NCESSCH) is a 12 character code where the first two characters indicate the state of the school, so the leading zero must be kept. That will be fixed and turned into the index. Also the column names will be made more explicit.

```python
#The NCESSCH column will be changed to type str and added the leading zero if missing.
tcs_sch_200506.NCESSCH = tcs_sch_200506.NCESSCH.astype(str).apply(lambda x: '0'+x 
                                                                  if len(x)<12 else x)

#Then, this column will be made index and the NCESSCH column is deleted.
tcs_sch_200506.index = tcs_sch_200506.NCESSCH
del tcs_sch_200506['NCESSCH']

#Renaming index and columns
tcs_sch_200506.index.name = 'NCES School ID'
tcs_sch_200506.columns = ['Teachers w/Bachelor Degree (%)', 'Teachers w/Masters or PhD (%)',
                          'Teachers w/1-5 yrs of exp (%)', 'Teachers w/5+ yrs of exp (%)',
                          'Mean yrs of teaching exp', 'Mean base salary for teachers w/1-5 yrs exp',
                          'Mean base salary for teachers w/5+ yrs exp', 'Mean base salary']
```

The documentation indicates the following about data values:
* 0: no occurence of the the data element
* -1: data is missing, either because it could not be measured or required reporting standard could not be met.
* -2: data are not applicable, meaning no value was expected to be measured.

As an example, let's consider a school where 80% of the teachers answered the survey. Of them, 75% of the teacher population reported to have either a Master's degree or a PhD. The other 5% either answered inconclusively about their education or there were not enough people with Bachelor's degrees to merit giving it a percentage (according to the documentation, the data must come from at least three full-time teachers). Furthermore, if less than 70% of the teacher population in a school were reported in the survey, the data for that school would be considered missing and have an entry of "-1". Therefore, these missing data will not be filled in as it would mean making up numbers about the schools.

```python
#changing all columns to float and replacing missing values
#The documentation indicates that -1 means missing data and -2 means not applicable,
#but we will consider both as NaN.
tcs_sch_200506 = tcs_sch_200506.astype(float)
tcs_sch_200506.replace([-1., -2.], np.nan, inplace=True)
```

```python
#Printing the tail to check for the result
tcs_sch_200506.tail()
```

The state of the school is included in the NCES School ID but it's difficult to read. To make it easier to read I added a State column and matched the first two characters of the NCES School ID with the [Federal Information Processing Standard (FIPS) state code](https://en.wikipedia.org/wiki/Federal_Information_Processing_Standard_state_code).

```python
#Matching states with FIPS code
fips_state_codes = {}
with open('fips_state_codes.txt', 'r') as file:
    line = file.readline()
    while line!='':
        line = line.split('\t')
        if len(line[1])==3:
            fips_state_codes[int(line[2])] = line[1].strip()
        line = file.readline()

#Adding the State column
states_list = []
for code in tcs_sch_200506.index.values:
    states_list.append(fips_state_codes[int(code[:2])])
tcs_sch_200506['State'] = states_list

#Reviewing the changes
tcs_sch_200506.sample(5)
```

```python
#Plotting all histograms of Mean Base Salary by State in the same figure, to notice scale
grouped_by_state0506 = tcs_sch_200506[['Mean base salary', 'State']].dropna().groupby('State')
fig_hist, ax_hist = plt.subplots()

for group in grouped_by_state0506:
    ax_hist.hist(group[1].values[:,0].astype('float'), label=group[0], alpha=0.5)
    plt.legend()
#
fig_hist.set_size_inches((12,8))
ax_hist.set_title('Mean Base Salary by State')
ax_hist.set_xlabel('Mean Base Salary (US$)')
ax_hist.set_ylabel('Frequency')
plt.show()
```

This shows there is a lot more schools from Florida in the database. The top 5 states with highest average 'Mean base Salary' are:

```python
description_by_state0506 = grouped_by_state0506.describe()
print(np.round(description_by_state0506['Mean base salary', 'mean'].sort_values(ascending=False)))
```

```python
#Saving to a csv file
tcs_sch_200506.to_csv('tcs_schoolyear_2005-06_v1a.csv')
```

## School Year 2006-07

This data set has too many states to present it in a single histogram. The states with the top 5 Mean Base Salaries will be listed.

```python
description_by_state = tcs_sch_200607[['Mean base salary', 'State']].dropna().groupby('State').describe()
print(np.round(description_by_state['Mean base salary','mean'].sort_values(ascending=False)[:5], 2))
```

```python
#Saving to a csv file
tcs_sch_200607.to_csv('tcs_schoolyear_2006-07_v1a.csv')
```
